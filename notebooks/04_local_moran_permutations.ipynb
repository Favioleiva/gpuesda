# Local Moran conditional permutations
Validate Local I, quadrant codes, pseudo-p values, and FDR on Columbus. Cross-backend p-values are statistically—not sequence—equivalent.

```python
import geopandas as gpd, libpysal, esda, numpy as np
from gpu_esda import MoranLocal, gpu_available
g = gpd.read_file(libpysal.examples.get_path('columbus.shp')); y = g.CRIME.to_numpy()
w = libpysal.weights.Queen.from_dataframe(g, use_index=True); w.transform='r'
ref = esda.Moran_Local(y,w,permutations=999,seed=12345)
cpu = MoranLocal(y,w.sparse,999,12345,'cpu')
print(np.max(np.abs(cpu.Is-ref.Is)), np.count_nonzero(cpu.q != ref.q), np.mean(np.abs(cpu.p_sim-ref.p_sim)))
if gpu_available():
    gpu = MoranLocal(y,w.sparse,999,12345,'gpu'); display(gpu.to_frame().head())
```
