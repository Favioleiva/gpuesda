# Panel scaling benchmark
Exercise observation × variable × period inputs and inspect the GPU break-even point.

```python
from gpu_esda import inverse_distance_weights, moran_panel, gpu_available
from gpu_esda.benchmarking import synthetic_coordinates, synthetic_panel, benchmark_call
records=[]
for n in [500,1000,1793]:
    c=synthetic_coordinates(n); panel=synthetic_panel(n,10,4); w=inverse_distance_weights(c,backend='cpu')
    for backend in (['cpu','gpu'] if gpu_available() else ['cpu']):
        records.append((n,backend,benchmark_call(lambda: moran_panel(panel,w,99,12345,backend),backend,3).to_dict()))
records
```
