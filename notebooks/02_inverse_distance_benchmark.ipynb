# Inverse-distance benchmark
Compare dense and CSR d^-2 construction, normalization, memory, and warmed lag.

```python
from gpu_esda import inverse_distance_weights, spatial_lag, gpu_available
from gpu_esda.benchmarking import synthetic_coordinates, benchmark_call
import numpy as np
coordinates = synthetic_coordinates(1793); y = np.random.default_rng(12345).normal(size=1793)
rows = []
for backend in (['cpu','gpu'] if gpu_available() else ['cpu']):
    for fmt in ['dense','csr']:
        w, d = inverse_distance_weights(coordinates, backend=backend, output_format=fmt, return_diagnostics=True)
        rows.append((backend, fmt, d.to_dict(), benchmark_call(lambda: spatial_lag(w,y,backend), backend).to_dict()))
rows
```
