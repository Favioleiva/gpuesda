# Open dataset validation
Load the official Columbus example and compare observed CPU/GPU statistics with PySAL.

```python
import geopandas as gpd, libpysal, esda, numpy as np
from gpu_esda import Moran, MoranLocal, gpu_available
gdf = gpd.read_file(libpysal.examples.get_path('columbus.shp'))
w = libpysal.weights.Queen.from_dataframe(gdf, use_index=True); w.transform = 'r'
y = gdf.CRIME.to_numpy()
reference = esda.Moran(y, w, permutations=0)
cpu = Moran(y, w.sparse, permutations=0, backend='cpu')
gpu = Moran(y, w.sparse, permutations=0, backend='gpu') if gpu_available() else None
reference.I, cpu.I, None if gpu is None else gpu.I
```
