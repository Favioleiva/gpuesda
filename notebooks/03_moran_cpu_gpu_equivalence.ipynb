# Moran CPU/GPU equivalence
Compare lag and Global Moran observed/permutation results with fixed backend seeds.

```python
import numpy as np
from gpu_esda import inverse_distance_weights, spatial_lag, Moran, gpu_available
from gpu_esda.benchmarking import synthetic_coordinates
c = synthetic_coordinates(500); y = np.random.default_rng(9).normal(size=500)
wc = inverse_distance_weights(c, backend='cpu', output_format='dense')
mc = Moran(y, wc, 999, 12345, 'cpu')
if gpu_available():
    wg = inverse_distance_weights(c, backend='gpu', output_format='dense')
    mg = Moran(y, wg, 999, 12345, 'gpu')
    print(np.max(np.abs(spatial_lag(wc,y,'cpu')-spatial_lag(wg,y,'gpu'))), mc.I, mg.I, mc.p_sim, mg.p_sim)
else: print(mc.summary())
```
